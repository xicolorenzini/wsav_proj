```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = 'https://github.com/xicolorenzini/wsav_proj/raw/refs/heads/main/wsav.xlsx'
```

```python
df = pd.read_excel(url, sheet_name='Planilha2')
df.head()
```

```python
df.info()
```

```python
means = df.groupby("Period").mean(numeric_only=True)

top5 = means.apply(lambda row: row.sort_values(ascending=False).head(5), axis=1)
top5.round(4)
```

```python
top5_rounded = top5.round(4)
df_heat = top5_rounded.copy()

fig = px.imshow(
    df_heat,
    color_continuous_scale="Spectral",
    labels=dict(x="L", y="Period", color="Mean"),
    aspect="auto",
    text_auto=True
)

# Define a cor de NaN como branco
fig.update_traces(
    zmin=df_heat.min().min(),
    zmax=df_heat.max().max(),
    hovertemplate="%{y} - %{x}: %{z}<extra></extra>",
    colorscale="Viridis",
    showscale=True,
    z=df_heat.values
)

# Fundo do grÃ¡fico em branco
fig.update_layout(
    plot_bgcolor="white",
    paper_bgcolor="white"
)

# Remove grades e linhas de fundo
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.show()
```

```python
!pip install dash
from dash import Dash, dcc, html, Input, Output, dash_table
from dash.dash_table.Format import Format, Scheme

# Lista de locais
L_options = ["L1", "L18", "L20", "L22", "L24", "L27", "L3", "L9"]
L_options = [l for l in L_options]

app = Dash(__name__)

app.layout = html.Div([
    html.H2("EstatÃ­sticas bÃ¡sicas para os valores de wsav do local"),

    # Dropdown para escolher qual L mostrar
    dcc.Dropdown(
        id="select-L",
        options=[{"label": l, "value": l} for l in L_options],
        value=L_options[0]
    ),

    dash_table.DataTable(
        id="table",
        columns=[
            {"name": "Period", "id": "Period"},
            {"name": "count", "id": "count"},
            {"name": "mean", "id": "mean", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "std", "id": "std", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "min", "id": "min", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "25%", "id": "25%", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "50%", "id": "50%", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "75%", "id": "75%", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
            {"name": "max", "id": "max", "type": "numeric", "format": Format(precision=4, scheme=Scheme.fixed)},
        ],
        page_size=10,
        style_table={"overflowX": "auto"},
        style_cell={"textAlign": "center", "padding": "5px", "fontSize": "20px", "fontFamily": "Helvetica"},
        style_header={"textAlign": "center", "fontWeight": "bold", "backgroundColor": "#f2f2f2"}
    )
])

@app.callback(
    Output("table", "data"),
    Input("select-L", "value")
)
def update_table(selected_L):
    # EstatÃ­sticas do local escolhido
    cols = [(selected_L, stat) for stat in ["count", "mean", "std", "min", "25%", "50%", "75%", "max"]]
    filtered = summary[cols].copy()
    filtered = filtered.reset_index()  # "Period" como coluna
    filtered.columns = ["Period", "count", "mean", "std", "min", "25%", "50%", "75%", "max"]
    return filtered.to_dict("records")

app.run(debug=True, mode='inline', port=8050)
```

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H2("GrÃ¡fico estilo boxplot para os valores de wsav do local"),
    html.P("x:"),
    dcc.Checklist(
        id='x-axis',
        options=['Period', 'W'],
        value=['Period'],
        inline=True
    ),
    html.P("y:"),
    dcc.RadioItems(
        id='y-axis',
        options=[{"label": l, "value": l} for l in L_options],
        value="L9",
        inline=True
    ),
    dcc.Graph(id="graph"),
])


@app.callback(
    Output("graph", "figure"),
    Input("x-axis", "value"),
    Input("y-axis", "value"))
def generate_chart(x, y):
    global df
    if isinstance(x, list) and len(x) > 0:
        x = x[0]
    fig = px.box(df, x=x, y=y, color=x)
    return fig


app.run(debug=True, mode='inline', port=8051)
```

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H2("GrÃ¡fico estilo scatter3D para os valores de wsav do local"),

    html.Div([
        html.Label("x:"),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in L_options],
            value='L9'
        )
    ], style={'width': '30%', 'display': 'inline-block'}),

    html.Div([
        html.Label("y:"),
        dcc.Dropdown(
            id='y-axis',
            options=['Period'],
            value='Period'
        )
    ], style={'width': '30%', 'display': 'inline-block', 'marginLeft': '2%'}),

    html.Div([
        html.Label("z:"),
        dcc.Dropdown(
            id='z-axis',
            options=['W'],
            value='W'
        )
    ], style={'width': '30%', 'display': 'inline-block', 'marginLeft': '2%'}),

    dcc.Graph(id='3d-scatter')
])

@app.callback(
    Output('3d-scatter', 'figure'),
    Input('x-axis', 'value'),
    Input('y-axis', 'value'),
    Input('z-axis', 'value')
)
def update_figure(x_col, y_col, z_col):
    fig = px.scatter_3d(df, x=x_col, y=y_col, z=z_col, color=z_col)
    fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))  # remove margens
    return fig

app.run(debug=True, mode='inline', port=8052)
```

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H2("GrÃ¡fico de dispersÃ£o com valores de wsav"),

    html.Label("x:"),
    dcc.Dropdown(
        id='x-axis',
        options=[{'label': col, 'value': col} for col in L_options],
        value='L18'
    ),

    html.Label("y:"),
    dcc.Dropdown(
        id='y-axis',
        options=[{'label': col, 'value': col} for col in L_options],
        value='L22'
    ),

    dcc.Graph(id='scatter-plot')
])

@app.callback(
    Output('scatter-plot', 'figure'),
    Input('x-axis', 'value'),
    Input('y-axis', 'value')
)
def update_graph(x_col, y_col):
    fig = px.scatter(
        df,
        x=x_col,
        y=y_col,
        hover_data=['W'],
        hover_name='Period',
        color='W',
        trendline='ols'
    )
    return fig


app.run(debug=True, mode='inline', port=8053)
```
